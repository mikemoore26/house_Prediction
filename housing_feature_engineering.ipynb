```python
def fill_missing_values(data):
    missing_f = [f for f in data.columns if data[f].isna().sum()>0]

    for f in missing_f:
        if data[f].dtype == "O":
            data[f] = data[f].fillna("Missing")
        else:
            data[f+"_na"] = np.where(data[f].isna(),1,0)
            data[f] = data[f].median()
    
    return data
            
```

# Convert Year to a more suitable number

```python
# data.columns
```

```python
year_f = [f for f in data.columns if "Year" in f or "Yr" in f]
year_f.remove('GarageYrBlt_na')
year_f
```

```python
def convert_year(data):
    year_f = [f for f in data.columns if "Year" in f or "Yr" in f]
    year_f.remove('GarageYrBlt_na')
    
    for f in year_f:
        if f != 'YrSold':
            data[f] = data["YrSold"] - data[f]
    return data
    
```

```python
def norm_out(data):
    outl = ['LotFrontage',"LotArea",'1stFlrSF','GrLivArea','SalePrice']
    
    for f in outl:
        if f == "SalePrice" and "SalePrice" in data.columns:
            data[f] = np.log(data[f])
        
    return data
```

```python
data = norm_out(data)
data
```

```python
cat_f = [f for f in data.columns if data[f].dtype == "O"]
for f in cat_f:
    print(f,'\n',data[f].unique())
```

```python
def cat_op(data):
    features = [f for f in data.columns if data[f].dtype == "O"]
    for f in features:
        if "SalePrice" in data.columns:
            temp = data.groupby(f)["SalePrice"].count()/len(data)
        else:
            temp = data.groupby(f).count()/len(data)

        temp_df = temp[temp > 0.01].index
        data[f] = np.where(data[f].isin(temp_df),data[f],"Rare_Var")
    
    return data
        
```

```python
data = cat_op(data)
for f in cat_f:
    print(f,'\n',data[f].unique())
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_label(data):
    feat = [f for f in data.columns if data[f].dtype == "O"]
    
    for f in feat:
        le = LabelEncoder()   
        data[f]= le.fit_transform(data[f])
        
    return data
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler,LabelEncoder
def scale_features(data):
    
    feat = [f for f in data.columns if f not in ["Id","SalePrice"]]
    
    scaler = MinMaxScaler()
    scaler.fit(data[feat])
    if "SalePrice" in data.columns:
        train_x = pd.concat([data[["Id","SalePrice"]].reset_index(drop=True),
                  pd.DataFrame(scaler.transform(data[feat]),columns=feat)],
                  axis=1)
    else:
        train_x = pd.concat([data[["Id"]].reset_index(drop=True),
                  pd.DataFrame(scaler.transform(data[feat]),columns=feat)],
                  axis=1)
    #print(train_x)
    return train_x
    
    
```

```python
def optimize_dataset(data):
    data = fill_missing_values(data)
    data = convert_year(data)
    data = norm_out(data)
    data = cat_op(data)
    data = encode_label(data)
    data = scale_features(data)
    return data
```
