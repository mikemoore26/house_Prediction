```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
data_train = pd.read_csv("data/train_data.csv")
data_test = pd.read_csv("data/test_data.csv")
train = data_train.copy()
test = data_train.copy()
```

```python
train.head()
```

```python
train_x = train.drop(['Id','SalePrice'],1)
train_y = train['SalePrice']
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
from sklearn import ensemble,metrics,model_selection
```

```python
classifer = ensemble.RandomForestClassifier(n_jobs=-1)
```

```python
param_grid = {
    "n_estimators": [100,200,300,400],
    "max_depth" : [1,3,5,7],
    "criterion" : ["gini","entropy"],
}
```

```python
model = model_selection.GridSearchCV(
    estimator = classifer,
    param_grid=param_grid,
    scoring="accuracy",
    verbose=10,
    n_jobs=1,
    cv=5,)
```

```python
model.fit(train_x,train_y)
print("best score"  .model.best_score_)
model.best_estimator_.get_params()
```

```python
param_grid = {
    "n_estimators": np.arange(100,1500,100),
    "max_depth" : np.arange(1,20),
    "criterion" : ["gini","entropy"],
}
```

```python
model = model_selection.RandomizedSearchCV(
    estimator = classifer,
    param_distributions=param_grid,
    n_iter=10,
    scoring="accuracy",
    verbose=10,
    n_jobs=1,
    cv=5,)
```

```python
model.fit(train_x,train_y)
print("best score",model.best_score_)
model.best_estimator_.get_params()
```

```python
feat_sel_model = SelectFromModel(Lasso(alpha=0.005,random_state=26))
feat_sel_model.fit(train_x,train_y)
```
